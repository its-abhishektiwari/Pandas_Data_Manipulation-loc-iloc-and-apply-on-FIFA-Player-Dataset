# 2. Apply + Lambda Function

```python
# use lmbda function to convert "height_cm" series to meters
df_players['height_cm'].apply(lambda x:x/100)

# alternative
df_players['height_cm']/100
```

```python
# use lambda function to convert "long_name" series to upper case
df_players['long_name'].apply(lambda name:name.upper())

# alternative with str attribute
df_players['long_name'].str.upper()
```

```python
df_players['dob'] = pd.to_datetime(df_players['dob'])

# use lambda function to get the year or "dob" series
df_players['dob'].apply(lambda birth:birth.year)

# alternative with dt attribute
df_players['dob'].dt.year
```

```python
# apply lambda function to dataframe in order to calculate bmi
df_players.apply(lambda x:x['weight_kg'] / ((x['height_cm']/100)**2),axis=1)
```

# 1. copy()

# 1.1 deep = True

```python
# deep=True by default (Modification to the data or indices of the copy
df_copy = df_players.copy()
```

```python
# update value in original dataframes
df_players.loc['L. Messi','height_cm'] = 190
```

```python
# copy vs original dataframes

df_players

df_copy
```

# 1.2 deep = False

```python
# deep=False (Any to the data of the original will be reflected in the shallow copy and vice versa)
df_copy_shallow = df_players.copy(deep=False)
```

```python
# update vlaue in original dataframe

df_players.loc['Cristiano Ronaldo','height_cm'] =200
```

```python
# shallow copy vs original dataframe
df_players
df_copy_shallow
```
